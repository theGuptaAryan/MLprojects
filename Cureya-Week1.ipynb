
Linear Regression


```python
data = pd.read_csv("whr.csv")
X = np.array(data[['Logged GDP per capita','Social support']])
y = np.array(data['Ladder score'])
X = preprocessing.scale(X)
```

```python
#chooses 20% of our data as testing after shuffling the entire data.
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train = X[:-20]
X_test = X[-20:]

# Split the targets into training/testing sets
y_train = y[:-20]
y_test = y[-20:]

#choose linear regression as classifier
clf = LinearRegression()
#clf = svm.SVR()
#clf = svm.SVR(kernel='poly')
clf.fit(X_train, y_train)
#accuracy = clf.score(X_test, y_test)
#print(accuracy) #only for classification problems for regression use mean sqaure error!

#for mean square error in regression problems.
y_pred = clf.predict(X_test)
# The coefficients
print('Coefficients: \n', clf.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))

plt.scatter(X_test[:,1], y_test,  color='black')
plt.scatter(X_test[:,1], y_pred, color='blue')

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# inear Regression model with a single feature Logged GDP per capita to predict Ladder score!!

X1 = np.array(data['Logged GDP per capita'])
X1_train = X1[:-20]
X1_train1= X1_train
X1_train = np.matrix(X1_train.reshape(len(X1_train), 1))
X1_test = X1[-20:]
X1_test1 = np.matrix(X1_test.reshape(len(X1_test), 1))
y_train = y[:-20]
y_test = y[-20:]
clf = LinearRegression()
clf.fit(X1_train, y_train)
#for mean square error in regression problems.
y1_pred = clf.predict(X1_test1)
y2_pred = clf.predict(X1_train)
# The coefficients
print('Coefficients: \n', clf.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y1_pred))

#plotting our hypothesis whcih is our trained data set!
plt.scatter(X1_train1,y_train, color='black')
plt.scatter(X1_train1, y2_pred, color='blue')

#plotting our testing data set!
plt.scatter(X1_test, y_test,  color='lime')
plt.scatter(X1_test, y1_pred, color='yellow')

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# Linera regression on randomised linear input as previous data form was biased!
X2 = np.array(data['Logged GDP per capita'])
X2 = np.matrix(X2.reshape(len(X2), 1))
y2 = np.array(data['Ladder score'])

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3)

clf2 = LinearRegression()
clf2.fit(X2_train, y2_train)

#predictions
y21_pred = clf2.predict(X2_test)
y22_pred = clf2.predict(X2_train)

# The coefficients
print('Coefficients: \n', clf2.coef_)

# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y2_test, y21_pred))


X2_trainer = X2_train.reshape(1,-1)
X2_trainer = X2_trainer.flatten()
#plotting our hypothesis whcih is our trained data set!
plt.scatter([X2_trainer[0]], y2_train, color='black')
plt.scatter([X2_trainer[0]], y22_pred, color='blue')


X2_tester = X2_test.reshape(1,-1)
X2_tester = X2_tester.flatten()
#plotting our testing data set!
plt.scatter([X2_tester[0]], y2_test,  color='lime')
plt.scatter([X2_tester[0]], y21_pred, color='yellow')

plt.show()
```
